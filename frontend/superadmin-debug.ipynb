```python
def scan_balance(text, max_line=None):
    paren = 0
    brace = 0
    in_single = False
    in_double = False
    in_back = False
    i = 0
    line = 1
    while i < len(text):
        c = text[i]
        if c == '\n':
            line += 1
            in_single = in_double = in_back = False
            i += 1
            continue
        if in_single:
            if c == "'":
                in_single = False
            elif c == '\\':
                i += 2
                continue
        elif in_double:
            if c == '"':
                in_double = False
            elif c == '\\':
                i += 2
                continue
        elif in_back:
            if c == '`':
                in_back = False
            elif c == '\\':
                i += 2
                continue
        else:
            if c == "'":
                in_single = True
            elif c == '"':
                in_double = True
            elif c == '`':
                in_back = True
            elif c == '/' and i+1 < len(text) and text[i+1] == '/':
                # skip to end of line
                while i < len(text) and text[i] != '\n': i += 1
                continue
            elif c == '/' and i+1 < len(text) and text[i+1] == '*':
                i += 2
                while i+1 < len(text) and not (text[i] == '*' and text[i+1] == '/'):
                    if text[i] == '\n': line += 1
                    i += 1
                i += 2
                continue
            elif c == '(':
                paren += 1
            elif c == ')':
                paren -= 1
            elif c == '{':
                brace += 1
            elif c == '}':
                brace -= 1
        if max_line and line > max_line:
            break
        i += 1
    return paren, brace, line

print(scan_balance(text, 465))
print('line 465:', lines[464])
print('line 466:', lines[465])
print('line 467 start', lines[466])
```
